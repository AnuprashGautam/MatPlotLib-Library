## tight_layout():
 When you create a plot with Matplotlib, sometimes the axis labels, titles, or other elements can overlap or get cut off, especially if you have multiple subplots. ` plt.tight_layout() ` adjusts the subplot parameters to make sure that all elements fit nicely in the figure area without overlapping or getting clipped.

You typically call `plt.tight_layout()` after you have created all your subplots but before you display or save the figure.


```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices=[60,40]

plt.pie(slices)

plt.title("My Pie Chart")
plt.tight_layout()

plt.show()
```

- Adding labels
- It doesn't need to be that the sum of all the proportions should be 100. Matplotlib will automatically calculate their proportions.

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices=[60,50]
labels=["sixty","fifty"]

plt.pie(slices,labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()

plt.show()
```

- Changing the edge color to make it easily differentiable.

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices=[60,40]
labels=["sixty","forty"]

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':"black"})

plt.title("My Pie Chart")
plt.tight_layout()

plt.show()
```

- Chaning the pies' colours. For it, we will have to make a similar list of colours to colour all the pies.
- we can also use the hex values to represent the colours.

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices=[60,40,20,10,15]
labels=["sixty","forty","twenty","ten","fifteen"]
colours=["red","green","orange","yellow","white"]

plt.pie(slices,labels=labels,colors=colours,wedgeprops={'edgecolor':"black"})

plt.title("My Pie Chart")
plt.tight_layout()

plt.show()
```

- <span style="color:red; text-decoration:underline;"><strong>explode</strong></span>: This is used to differentiate a specific pie from the whole pie chart. The value is defined between `0.1` to `1.0` representing the amount by which the pie will be shifted wrt radius.

- <span style="color:red; text-decoration:underline;"><strong>shadow</strong></span>: Provide a shadow i.e. 3D effect.

- <span style="color:red; text-decoration:underline;"><strong>startangle</strong></span>: The firts category of the pie chart starts at the angle defined in `startangle=`.
- <span style="color:red; text-decoration:underline;"><strong>autopct</strong></span>: This arguement is used to mention the percentage of each pie taking the amount of the whole pie chart.
e.g. `autopct="%1.1f%%"`

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
