## Numpy in Matplotlib

### arange() function: 

- The `arange` function in NumPy is used to create arrays with regularly spaced values. You specify the start, stop, and step size, and it generates numbers accordingly. For example, `np.arange(0, 10, 2)` will create an array starting from 0, ending before 10, with steps of 2, resulting in `[0, 2, 4, 6, 8]`. It's handy for creating sequences of numbers quickly and efficiently.

````python
import numpy as np
from matplotlib import pyplot as plt

x_dev=[10,20,30,40,50,60,70,80]
x_indices=np.arange(len(x_dev))
width=0.25

y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]
y_jsdev=[113,221,311,411,521,607,705,802]

plt.bar(x_indices-width,y_dev,width=width,label="Developers")
plt.bar(x_indices,y_pydev,width=width,label="Python Developers")
plt.bar(x_indices+width,y_jsdev,width=width,label="Java Script Developers")

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
````

## 1. Read the csv file from standard library.

* Counter class: In Python, the `collections` module provides the `Counter` class, which is useful for counting the occurrences of elements in an iterable (like a list or a string). The `Counter` class creates a dictionary-like object where elements are stored as keys and their counts as values. This can be handy for tasks like finding the most common elements in a list or determining the frequency of characters in a string. It's a convenient tool for basic data analysis and manipulation.

``` python
from collections import Counter
# Create a list of fruits
fruits = ['apple', 'banana', 'orange', 'apple', 'banana', 'apple']
# Use Counter to count the occurrences of each fruit
fruit_counter = Counter(fruits)
# Display the result
print("Count of each fruit:")
for fruit, count in fruit_counter.items():
    print(f"{fruit}: {count}")

````python
import csv
import numpy as np
from collections import Counter
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

with open ("data.csv") as csv_file:
    csv_reader=csv.DictReader(csv_file)

    language_counter=Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages=[]
popularity=[]

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

plt.bar(languages,popularity,label="Realtion between language and its popularity")

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Most Popular Programming Languages")
````

## 2. Reading the file using pandas

````python
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

data=pd.read_csv("data.csv")
ids=data["Responder_id"]
lang_responses=data["LanguagesWorkedWith"]

language_counter=Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages=[]
popularity=[]

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()


plt.barh(languages,popularity,label="Realtion between language and its popularity")

plt.xlabel("Popularity")
plt.title("Most Popular Programming Languages")
````
