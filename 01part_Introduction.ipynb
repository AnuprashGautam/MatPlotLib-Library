# Introduction
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It provides a wide range of plotting tools and functionalities for generating various types of plots such as line plots, scatter plots, bar plots, histograms, and more. 

## Printing the single relation.
- show() function is required to print the plot.

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]

plt.plot(x_dev,y_dev)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
```

## Printing the double relation in a single graph

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]

plt.plot(x_dev,y_dev)
plt.plot(x_dev,y_pydev)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
```

## Legend
It is used to give the name to the line showing the relatin between the axes in the graph.
There are two ways to add the Legends:
1. First is to add at the last in which we have to keep in mind the serial of the plots.

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]

plt.plot(x_dev,y_dev)
plt.plot(x_dev,y_pydev)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.legend(["Developers","Python Developers"])
plt.show()
```

2. In the second method, we don't need to keep in mind the serial of the plots.

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]

plt.plot(x_dev,y_dev,label="Developers")
plt.plot(x_dev,y_pydev,label="Python Developers")

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
```

## Formating the plot

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]
y_jsdev=[113,221,311,411,521,607,705,802]

plt.plot(x_dev,y_dev,marker="o",color="g",label="Developers")
plt.plot(x_dev,y_pydev,marker=".",linestyle="--",color="b",label="Python Developers")
plt.plot(x_dev,y_jsdev,marker="*",linestyle="-.",color="y",label="Java Script Developers")

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
```

## linewidth
- By default width is 1.

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]
y_jsdev=[113,221,311,411,521,607,705,802]

plt.plot(x_dev,y_dev,linewidth=3,marker="o",color="g",label="Developers")
plt.plot(x_dev,y_pydev,linewidth=2,marker=".",linestyle="--",color="b",label="Python Developers")
plt.plot(x_dev,y_jsdev,marker="*",linestyle="-.",color="y",label="Java Script Developers")

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
```

## Grid

```python
from matplotlib import pyplot as plt

x_dev=[1,2,3,4,5,6,7,8]
y_dev=[100,203,342,452,545,633,733,785]
y_pydev=[123,250,345,488,573,599,732,800]
y_jsdev=[113,221,311,411,521,607,705,802]

plt.plot(x_dev,y_dev,linewidth=3,marker="o",color="g",label="Developers")
plt.plot(x_dev,y_pydev,linewidth=2,marker=".",linestyle="--",color="b",label="Python Developers")
plt.plot(x_dev,y_jsdev,marker="*",linestyle="-.",color="y",label="Java Script Developers")

plt.grid(True)
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")
plt.show()
```
