- Line plots are a type of graph that show data points connected by straight lines. They're great for displaying trends over time or relationships between variables. Each data point represents a value, and the lines between them help visualize patterns or changes.

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('newdata.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

- <span style="color:Yellow; text-decoration:underline;"><strong>fill_between</strong></span>: In Matplotlib, the `fill_between` function is used to fill the area between two lines in a line chart. It's handy for highlighting certain ranges or showing the difference between two lines. 
For example, you can fill the area between a line and the x-axis.

  The `alpha` attribute in Matplotlib is used to control the transparency of a plot element, such as lines, markers, or filled areas. It ranges from 0 to 1, where 0 is completely transparent (invisible) and 1 is fully opaque (solid). Adjusting the alpha value can help emphasize certain parts of a plot or make it easier to distinguish overlapping elements.

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('newdata.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages,py_salaries,alpha=0.25)

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

- Adding a median for dividing the plot into two distinguisable parts.
- We can even add the conditions to it.

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('newdata.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages,py_salaries,overall_median,alpha=0.25)



plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

- <span style="color:Yellow; text-decoration:underline;"><strong>interpolate</strong></span>: In the `fill_between` method of Matplotlib, the `interpolate` parameter is an optional argument that determines whether the filled region should be interpolated between data points. 

   When set to `True`, it interpolates the fill between data points, resulting in a smoother filled area. 
  
   When set to `False`, it fills the area using straight lines between data points. 
 
   This parameter is particularly useful when you have sparse data points and want to create a smoother filled region between them.

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('newdata.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages,py_salaries,overall_median,
                 where=(py_salaries>overall_median),color="green", interpolate=True,alpha=0.25)

plt.fill_between(ages,py_salaries,overall_median,
                 where=(py_salaries<=overall_median),color="red", interpolate=True,alpha=0.25)



plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

- We can also add the legends for the coloured regions too.

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('newdata.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
