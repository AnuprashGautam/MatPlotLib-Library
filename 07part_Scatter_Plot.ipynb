- A scatter plot is a type of graph where each individual data point is represented by a dot. The position of the dot on the graph corresponds to the values of two variables. It helps us see if there's a relationship between the two variablesâ€”like if they move together, move in opposite directions, or don't seem to have any connection at all. It's great for spotting patterns and outliers in data.

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use("seaborn")

X=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
Y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

plt.scatter(X,Y)
```

- `s` attribute is used to set the size of dots.
- `marker` is to select the style of marker.
- `c` is to select the colour of marker.
- `alpha` is to select the transparency of the marker.
- `edgewidth` to set the width of edge of the marker.
- `edgecolor` to set the color of the edge of the marker.

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use("seaborn")

X=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
Y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

plt.scatter(X,Y,s=100,marker="X",c="green",alpha=0.75,linewidth=2,edgecolors="red")
```

- `cmap` it is called colour map. It basically represent the shades of a specific colour.Its by default value is `Grey`.

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use("seaborn")

X=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
Y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

colors=[7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]

plt.scatter(X,Y,s=100,c=colors,cmap="Greens",alpha=0.75,linewidth=2,edgecolors="red")
```

## Color Bar

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use("seaborn")

X = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
Y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

# Creating the scatter plot
scatter = plt.scatter(X, Y, s=sizes, c=colors, cmap="Greens", alpha=0.75, linewidth=2, edgecolors="red")

# Adding colorbar
cbar = plt.colorbar(scatter)
cbar.set_label("Satisfaction Level")

plt.show()
```

## Working om Real World Data

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
